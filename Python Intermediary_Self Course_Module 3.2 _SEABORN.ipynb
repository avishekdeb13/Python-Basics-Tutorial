## SEABORN - It provides high level interface for drawing attractive and informative statistical graphs

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
datafile = pd.read_csv("E:/ONLINE COURSES/Side projects/MY PROJECT/Datasets/Toyota.csv",
                    index_col= 0,
                    na_values= ["??", "????"])
```

```python
df = datafile
```

```python
df
```

```python
df.info()
```

```python
df.dropna(inplace = True) # Removing all missing values
```

```python
df.isnull().sum()
```

## 1. SCATTER PLOT

```python
sns.set(style = "darkgrid") # style = white/ dark/whitegrid/ darkgrid/ ticks  can be used
sns.scatterplot(x = "Age", y = "Price", data = df)
```

```python
sns.scatterplot(x = "Age", y = "Price", data = df, color = "red")
```

```python
# To include a third variable in the scatter plot using different colour.
sns.set(style = "whitegrid")
sns.scatterplot(x = "Age", y = "Price", data = df, hue = "FuelType")
```

```python
# To include a third variable in the scatter plot using different colour and different marker.
sns.set(style = "whitegrid")
sns.scatterplot(x = "Age", y = "Price", data = df, hue = "FuelType", style = "FuelType")
```

```python
plt.figure(figsize = (20,10))
sns.scatterplot(x = "Age", y = "Price", data = df, hue = "FuelType", size="FuelType", sizes = (200,100))
```

## 2. LINEPLOT - To show the relationship between two variables with a line

```python
sns.lineplot(df.Age, df.Price)
```

```python
plt.figure(figsize = (20,10))
sns.lineplot(df.Age, df.Price)
plt.show()
```

## 3.REGRESSION PLOT

```python
sns.regplot(df.Age, df.Price)
plt.show()
```

```python
# To remove the regression line.
plt.figure(figsize = (20,10))
sns.regplot(df.Age, df.Price, marker = "*", fit_reg =  False )
plt.show()
```

## 4. LMPLOT - Plot data and regression model fits across a FacetGrid.

```python
sns.lmplot(x = "Age", y = "Price",data = df, fit_reg = False, hue = "FuelType", legend = True, palette = "Set1", size = 10)
```

## 5.HISTOGRAM

```python
plt.figure(figsize = (20,10))
sns.distplot(df.Price)
plt.show()
```

```python
# To remove the line and increase the bins
plt.figure(figsize = (20,10))
sns.distplot(df.Price, kde = False, bins = 40)
plt.show()
```

## 7.BARPLOT

```python
plt.figure(figsize = (10,5))
sns.countplot(df.FuelType)
plt.show()
```

## 8. GROUPED BARPLOT

```python
sns.set(style = "darkgrid")
plt.figure(figsize = (10,5))
sns.countplot(x ="FuelType",data = df, hue = "Automatic")
plt.show()
```

## 9. BOX AND WHISKER PLOT -To visually interpret the  number summary

```python
# Only numerical variable
plt.figure(figsize = (10,5))
sns.boxplot(y = df["Price"])
```

```python
# Numerical vs. Categorical variable
plt.figure(figsize = (10,5))
plt.title("Price vs. FuelType")
sns.boxplot(x = df["FuelType"], y = df["Price"])
plt.show()
```

```python
# Numerical vs. 2 Categorical variables
plt.figure(figsize = (10,5))
plt.title("Price vs. FuelType and Automatic")
sns.boxplot(x = df["FuelType"], y = df["Price"], hue = df["Automatic"])
plt.show()
```

```python
# Box whiskers plot and histogram
f, (ax_box, ax_hist) = plt.subplots(2, gridspec_kw = {"height_ratios": (.15, .85)})

sns.boxplot(df["Price"], ax = ax_box)
sns.distplot(df["Price"], ax = ax_hist, kde = False)
```

## 10. PAIRWISE PLOT - To show pairwise relationship in the data

```python
sns.pairplot(data = df)
plt.show()
```

## CATPLOT- To show relationship between a numerical variable and one or more categories

```python
sns.catplot(x = "Price", y = "FuelType", data = df)
plt.show()
```

```python
sns.catplot(x = "Price", y = "FuelType", hue="Automatic", data = df)
plt.show()
```

```python
plt.figure(figsize = (15,7))
sns.heatmap(df.corr(),annot=True)
```
