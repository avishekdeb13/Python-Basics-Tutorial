### Relation between variables

```python
# One way table - To compute a cross tabulation of 1,2, or more variables
pd.crosstab(index = df["FuelType"], columns = "counts")
```

```python
# 2 way table: To look at the frequency distribution of gearbox type w.r.t. to fuel type of the car.
pd.crosstab(index = df["Automatic"], columns = df["FuelType"], 
            dropna = True )
```

```python
# Joint Probability : The occurance of two independent event occuring at the same time.
pd.crosstab(index = df["Automatic"], columns = df["FuelType"],
            normalize=True,
            dropna = True )
```

```python
# Marginal probability - The occurance of a single event.
pd.crosstab(index = df["Automatic"], columns = df["FuelType"],
            normalize=True,
            margins = True,
            dropna = True )
# Interpretation - The probability of having manual gearbox when the car type is CNG or Diesel Or Petrol is 95% 
```

```python
# Conditional probabilities - The prob. of an event A considering anothe event B  has already occured.
pd.crosstab(index = df["Automatic"], columns = df["FuelType"],
            margins = True,
            normalize='index',
            dropna = True )
# Interpretation : Given the gearbox is manual, the prob. of getting a CNG fueltype is 1%, Diesel is 10% and petrol is 88%
# Row sum is 1
```

```python
# Conditional probabilities - The column sum is 1 here
pd.crosstab(index = df["Automatic"], columns = df["FuelType"],
            margins = True,
            normalize='columns',
            dropna = True )
# Interpretation - Given that the Fueltype is CNG, the prob. of having manual gearbox is 100%
# Given the Fueltype is Petrol, manual 94% and automatic is 6%
```
