### IMPUTATION

```python
## Index name change (2 ways)
df = df.rename({"A": "A_name_changed"})
df.index.values[1] = "B_name_changed"

##Column name change (2 ways)
df = df.rename({"Advt": "ADVERTISEMENT_name_changed"}, axis = 1)
df.columns.values[1] = "SALES_name_changed"

##New columns addition in the dataset
df = df.reindex(columns = list(df.columns) + ["new_col1"])
df = df.reindex(columns = list(df.columns) + ["new_col2"])

##New column value fill.
df.iloc[:,2] = 0
df.iloc[:,2] = 1

##Removal of columns (2 ways)
df.drop(["new_col1"], axis= 1, inplace = True)
df.drop(df.columns[2], axis = 1, inplace = True)
```

```python
## Lets create a new columns which is the addition of sales and advertisement and insert in the existing dataset
add = df.ADVERTISEMENT_name_changed+df.SALES_name_changed

# Impute the new columns values to the existing dataset (3 ways)
df = df.reindex(columns = list(df.columns) + ["new_col1"]) # Creating an empty column

df = df.assign(new_col1 = add) # 1st way
df.loc[:, "new_col1"] = add    # 2nd way
df.insert(2,"new_col1", add)   # 3rd way

```

```python
## Data sorting
df[df.SALES_name_changed>1100]
```

```python
## A new column with string values based on certain condition
df.insert(3, "new_col_summation", " ")
```

```python
for i in range(0, len(df.new_col1)):
    if(df.new_col1[i] > 1150):
        df["new_col_summation"][i] = "High"
    else:
        df["new_col_summation"][i] = "Low"
```
