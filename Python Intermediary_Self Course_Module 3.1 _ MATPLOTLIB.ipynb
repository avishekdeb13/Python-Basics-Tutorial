## MTAPLOTLIB  - to create 2D Graphs and plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datafile = pd.read_csv("E:/ONLINE COURSES/Side projects/MY PROJECT/Datasets/Toyota.csv",
                    index_col= 0,
                    na_values= ["??", "????"])
```

```python
df = datafile
```

```python
df.info()
```

```python
df.dropna(inplace = True) # Removing all missing values
```

## 1.SCATTER PLOT - Used to show the relationship between two numerical variables, correlation graph.

```python
plt.scatter(df["Age"], df["Price"])
# Interpretation - Thus from the scatter plot it can be concluded that the price of the car decreases as the age of the car increases.
# Inverse relationship between the price and age variable.
```

```python
plt.scatter(df["Age"], df["Price"],s = 200, c="g")
```

```python
plt.scatter(df["Age"], df["Price"],
           c = "green",
           marker = "+")
```

```python
plt.scatter(df.Age, df.Price,
           c = "orange",
           marker = ">",
           )
plt.title("Scatter plot to show the relationship between age and price ")
```

```python
plt.scatter(df.Age, df.Price,
           c = "black",
           marker = "*",
           )
plt.title("Scatter plot to show the relationship between age and price ")
plt.xlabel("Age of car")
plt.ylabel("Price of car")
```

```python
plt.scatter(df.Age, df.Price, s = 100,
           c = "skyblue",
           marker = "*",
           )
plt.title("Scatter plot to show the relationship between age and price ")
plt.xlabel("Age of car")
plt.ylabel("Price of car")
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.scatter(df.Age, df.Price,
           c = "skyblue",
           marker = "*",
           s = 500 ,
           )
plt.title("Scatter plot to show the relationship between age and price ")
plt.xlabel("Age of car")
plt.ylabel("Price of car")
plt.show()
```

## 2. HISTOGRAM - Frequency distribution of a continuous variable(numeric variable)

```python
plt.hist(df.KM)
plt.show()
```

```python
plt.hist(df.KM,color= "red")
plt.show()
```

```python
plt.hist(df.KM, color = "yellow", edgecolor = "blue")
plt.show()
```

```python
plt.hist(df.KM, color = "blue", edgecolor = "white", bins = 20)
plt.show()
```

```python
plt.figure(figsize = (20,10))
plt.hist(df.KM, color = "lightblue", edgecolor = "black", bins = 20)
plt.title("Histogram of KM")
plt.xlabel("KM")
plt.ylabel("Counts")
plt.show()
```

## 3. BARPLOT - Frequency distribution of a categorical variable (Thus there will be space in between the bars unlike histogram)

```python
pd.crosstab(df.FuelType, "counts")
```

```python
counts = [968, 116, 12]
FuelType = ("Petrol","Diesel", "CNG")
index = np.arange(len(FuelType))

plt.bar(x = index,height = counts, color = ["red", "blue", "green"])
```

```python
counts = [968, 116, 12]
FuelType = ("Petrol","Diesel", "CNG")
index = np.arange(len(FuelType))

plt.bar(x = index,height = counts, color = ["red", "blue", "green"])

plt.xticks(index, FuelType, rotation = 90)
plt.show()
```
